## Level 2 

```python
msft1=web.get_data_yahoo("MSFT",start="2019-06-06",end="2020-06-06")
msft2=web.get_data_yahoo("MSFT",start="2020-05-06",end="2020-06-06")
```

```python
fig,ax=plt.subplots(figsize=(15,5))
ax.plot(msft1.Close)
ax.set_xlabel("Date")
ax.set_ylabel("Close")
ax.plot(msft2['Close'])
ax.set_title("MSFT")

ax2=ax.twinx()
ax2.set_ylabel("Last 1 month",fontsize=15)
plt.show()
# save the plot as a file
fig.savefig('msft.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
